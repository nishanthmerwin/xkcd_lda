```python
from matplotlib import pyplot as plt
import json
import datetime
```

```python
with open("../data/all_comics.json") as fp:
    comics = json.load(fp)

print(json.dumps(comics[0], indent=4))
```

```python
start_year = 2006
end_year = 2019
start_month = 1
end_month = 12

counts = []

for year in range(start_year, end_year+1):
    for month in range(start_month, end_month+1):
        if year == 2019 and month > 1:
            break
        matched_comics = len([x for x in comics if 
                              x['month'] == str(month) and 
                              x['year'] == str(year)])
        print(matched_comics)
        counts.append(dict(month=month, year=year, num_comics=matched_comics))

for data in counts:
    data['date'] = datetime.datetime(data['year'], data['month'], 1)
```

```python



dates = [x['date'] for x in counts]
num_comics = [x['num_comics'] for x in counts]


plt.xkcd()
fig, ax = plt.subplots(1,1,figsize=(8,5))
ax.plot(dates, num_comics)
ax.set_xlabel("Time")
ax.set_ylabel("Number of comics per month")

plt.savefig("../figs/productivity.svg")
plt.show()
```
